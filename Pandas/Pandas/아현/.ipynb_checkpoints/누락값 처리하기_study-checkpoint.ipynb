## ÎàÑÎùΩÍ∞í Ï≤òÎ¶¨ÌïòÍ∏∞

- fillna() : Ï≤òÎ¶¨Ìï¥Ïïº ÌïòÎäî df ÌÅ¨Í∏∞Í∞Ä Îß§Ïö∞ ÌÅ¨Í≥† Î©îÎ™®Î¶¨Î•º Ìö®Ïú®Ï†ÅÏúºÎ°ú ÏÇ¨Ïö©Ìï¥ÏïºÌïòÎäî Í≤ΩÏö∞Ïóê ÏûêÏ£º Ïì∞ÏûÑ

```python
ebola
```

```python
print(ebola.fillna(0))
```

- method='ffill' : ÎàÑÎùΩÍ∞íÏù¥ ÎÇòÌÉÄÎÇòÍ∏∞ Ï†ÑÏùò Í∞íÏúºÎ°ú ÎàÑÎùΩÍ∞íÏù¥ Î≥ÄÍ≤Ω

```python
ebola.iloc[0:10, 0:5]
```

```python
print(ebola.fillna(method="ffill").iloc[0:10, 0:5])
```

- method='bfill' : ÎàÑÎùΩÍ∞íÏù¥ ÎÇòÌÉÄÎÇú Ïù¥ÌõÑÏùò Ï≤´ Î≤àÏß∏ Í∞íÏúºÎ°ú ÎàÑÎùΩÍ∞íÏù¥ Î≥ÄÍ≤Ω

```python
ebola.iloc[0:10, 0:5]
```

```python
print(ebola.fillna(method="bfill").iloc[0:10, 0:5])
```

- interpolate() : ÎàÑÎùΩÍ∞í ÏñëÏ™ΩÏóê ÏûàÎäî Í∞íÏùÑ Ïù¥Ïö©Ìï¥ Ï§ëÍ∞ÑÍ∞íÏùÑ Íµ¨ÌïúÌõÑ Ïù¥Î•º ÌÜµÌï¥ ÎàÑÎùΩÍ∞íÏùÑ Ï≤òÎ¶¨

```python
ebola.iloc[0:10, 0:5]
```

```python
print(ebola.interpolate().iloc[0:10, 0:5])
```
