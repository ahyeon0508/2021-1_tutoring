# 그룹 연산

# 1. 집계

- gapminder.tsv 파일 불러와서 year열을 기준으로 데이터를 그룹화한 다음 lifeExp 열의 평균을 구하세요

```python
import pandas as pd

df = pd.read_csv('data/gapminder.tsv', sep = "\t")
```

```python
avg_life_exp_by_year = df.groupby('year')['lifeExp'].mean()
avg_life_exp_by_year
```

## 평균값을 구하는 사용자 함수와 groupby 메서드

```python
def my_mean(values):
    n = len(values)
    
    sum = 0 
    for value in values:
        sum += value
    
    return sum / n
```

```python
agg_my_mean = df.groupby('year')['lifeExp'].agg(my_mean)
print(agg_my_mean)
```

## 집계 메서드를 리스트, 딕셔너리에 담아 전달하기

```python
import numpy as np
gdf = df.groupby('year').lifeExp.agg([np.count_nonzero, np.mean, np.std]) 
print(gdf)
```

```python
gdf_dict = df.groupby('year').agg({'lifeExp': 'mean', 'pop': 'median', 'gdpPercap': 'median'}) 
print(gdf_dict)
```

# 2. 변환

## 표준점수
- 데이터의 평균값 0, 표준편차 1로 변환
- outlier 제거

```python
def my_zscore(x):
    return (x - x.mean()) / x.std()
```

```python
transform_z = df.groupby('year')['lifeExp'].transform(my_zscore)
print(transform_z.head())
```

## 누락값을 평균값으로 처리하기

```python
import seaborn as sns 
import numpy as np

np.random.seed(42)
tips_10 = sns.load_dataset('tips').sample(10)
tips_10.loc[np.random.permutation(tips_10.index)[:4], 'total_bill'] = np.NaN # 임의로 4개 누락값으로 변경

print(tips_10)
```

- 남성의 데이터가 많음 -> 여성과 남성을 구분하여 total_bill 열의 평균값으로 채워야함

```python
count_sex = tips_10.groupby('sex').count()
print(count_sex)
```

```python
def fill_na_mean(x):
    avg = x.mean() 
    return x.fillna(avg)
```

```python
total_bill_group_mean = tips_10.groupby('sex')['total_bill'].transform(fill_na_mean)
tips_10['fill_total_bill'] = total_bill_group_mean # 새로운 값 추가
print(tips_10)
```

# 4. 그룹 오브젝트
- groupby 메서드가 반환하는 값

```python
tips_10 = sns.load_dataset('tips').sample(10, random_state=42) 
print(tips_10)
```

```python
grouped = tips_10.groupby('sex')
print(grouped)
```

```python
print(grouped.groups)
```

## 그룹 오브젝트의 평균 구하기
