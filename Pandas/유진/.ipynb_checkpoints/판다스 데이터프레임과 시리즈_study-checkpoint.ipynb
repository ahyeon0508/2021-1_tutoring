## to_datetime ì´ìš©í•´ì„œ datetimeí˜•ìœ¼ë¡œ ë°”ê¾¸ê¸°
- pd.to_date(data, format=í˜•ì‹)

```python
born_datetime = pd.to_datetime(scientists['Born'], format="%Y-%m-%d")
born_datetime
```

```python
died_datetime = pd.to_datetime(scientists['Died'], format="%Y-%m-%d")
print(died_datetime)
```

```python
scientists['born_dt'], scientists['died_dt'] = born_datetime, died_datetime
print(scientists.head())
```

```python
scientists['age_days_dt'] = (scientists['died_dt'] - scientists['born_dt']) # datetimeì€ ê³„ì‚° ê°€ëŠ¥
print(scientists)
```
