# apply

- 사용자가 작성한 함수를 한 번에 데이터프레임의 각 행과 열에 적용하여 실행할 수 있게 해주는 메소드
- 대용량의 데이터를 처리할 때는 for문보다 빠름

## 제곱 함수와 n 제곱 함수 만들기

```python
def my_sq(n):
    return n ** 2
```

```python
def my_exp(x,n):
    return x ** n
```

## apply 적용

```python
import pandas as pd

df = pd.DataFrame({'a': [10, 20, 30], 'b': [20, 30, 40]})

df
```

- df['a']의 값이 첫 번째 인자로 넘어감

```python
sq = df['a'].apply(my_sq)
sq
```

```python
ex = df['a'].apply(my_exp, n=2)
ex
```

```python
def avg_3(x, y, z):
    return (x + y + z) / 3
```

```python
#print(df.apply(avg_3))
```

```python
def avg_3_apply(col):
    x = col[0]
    y = col[1]
    z = col[2]
    return (x + y + z) / 3

print(df.apply(avg_3_apply))
```

```python
def avg_2_apply(row):
    sum = 0
    for item in row:
        sum += item
    return sum / df.shape[1]
```

```python
print(df.apply(avg_2_apply, axis = 1)) #1=행, 0=열
```
