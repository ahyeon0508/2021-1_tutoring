# 1. 라이브러리 호출 및 데이터 읽기

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder #한글을 숫자로 바꿔줌
from sklearn.compose import make_column_transformer #신기한 거
import numpy as np #테스트 데이터 추가 시 사용
```

```python
train_df = pd.read_excel('https://github.com/cranberryai/todak_todak_python/blob/master/machine_learning/regression/carprice_E1SUl6b.xlsx?raw=true', sheet_name='train')
test_df = pd.read_excel('https://github.com/cranberryai/todak_todak_python/blob/master/machine_learning/regression/carprice_E1SUl6b.xlsx?raw=true', sheet_name='test')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
x_train = train_df.drop(['가격'], axis=1)
x_test = test_df.drop(['가격'], axis=1)
y_train = train_df['가격']
y_test = test_df['가격']
```

```python
x_train.head()
```

# 3. 데이터 전처리

```python
transformer = make_column_transformer(
        (OneHotEncoder(),['종류','연료','변속기']), #원핫인코더로 한번에 가능
        remainder = 'passthrough') #튜플에서 정의하지 않은 변수들은 그대로 살려 놓음
```

```python
x_train = transformer.fit_transform(x_train) #
x_test = transformer.transform(x_test) #train이 되어서 자동으로 test도 fit 해줌 그래서 그냥 transform 사용해도 됨
```

```python
x_train
```

# 4.데이터 모델링

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
print(model.score(x_train, y_train))
print(model.score(x_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor(random_state=43)
```

```python
forest.fit(x_train, y_train)
```

```python
print(forest.score(x_train, y_train))
print(forest.score(x_test, y_test))
```
