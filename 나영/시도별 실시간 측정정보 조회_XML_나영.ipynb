```python
from bs4 import BeautifulSoup #HTML 문서를 탐색해서 원하는 부분만 쉽게 뽑아주는 라이브러리

def util(url) :
    request = urllib.request.urlopen(url)
    xml = request.read()
    soup = BeautifulSoup(xml, "html.parser")
    
    return soup
#우리가 읽기 쉽도록 바꿔줌
```
