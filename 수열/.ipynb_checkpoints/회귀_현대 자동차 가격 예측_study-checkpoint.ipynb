# 1. 라이브러리 호출 및 데이터 읽기

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder # 데이터를 숫자로 바꿔줌 ex) 야구데이터 숫자로 바꾸기 
from sklearn.compose import make_column_transformer # 
import numpy as np
```

```python
train_df = pd.read_excel('https://github.com/cranberryai/todak_todak_python/blob/master/machine_learning/regression/carprice_E1SUl6b.xlsx?raw=true', sheet_name='train')
test_df = pd.read_excel('https://github.com/cranberryai/todak_todak_python/blob/master/machine_learning/regression/carprice_E1SUl6b.xlsx?raw=true', sheet_name='test')
```

```python
train_df.head()
```

```python
x_train = train_df.drop(['가격'], axis = 1)
x_test = test_df.drop(['가격'], axis = 1)
```

```python
y_train = train_df['가격']
y_test = test_df['가격']
```

```python
x_train.head()
```

```python
len(x_train)
```

# 2. 데이터 시각화

# 3. 원핫인코딩 : 데이터전처리과정(한글 데이터 바꾸기) 

```python
transformer = make_column_transformer((OneHotEncoder(),['종류','연료','변속기']),
                                      remainder = 'passthrough') #튜플에서 정의하지 않은 변수들은 그대로 살려놓음
```

```python
x_train = transformer.fit_transform(x_train) # train에서 fit 쓰기
x_test = transformer.transform(x_test) # test에서는 train에서 fit된 것을 그대로 이어받아서 transform만 씀 
```

```python
x_train
```

# 4. 데이터 모델링

## 1. linear regression
## 2. randomforest

```python
from sklearn.linear_model import LinearRegression
```

```python
model =  LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
print(model.score(x_train, y_train))
print(model.score(x_test, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
forest = RandomForestRegressor(random_state = 43) # randomforest는 다 다르게 나옴
forest.fit(x_train, y_train)
```

```python
print(forest.score(x_train, y_train))
print(forest.score(x_test, y_test))
```
